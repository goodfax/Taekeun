# 한국어 형태소 분석<br><br>

#### Konlpy: 꼬꼬마, 코모란, 트위터
    https://konlpy-ko.readthedocs.io/ko/v0.5.2/morph/#pos-tagging-with-konlpy
    - 한국어 정보처리를 위한 파이썬 패키지
    - 사전기반
        * Windows에서 Konlpy 설치로는 은전한닢 사용 불가

#### eunjeon: 은전한닢
    https://github.com/koshort/pyeunjeon
    - 은전한닢 프로젝트와 mecab 기반의 한국어 형태소 분석기의 독립형 python 인터페이스
    - 사전 + 딥러닝 기반

<br>
#### * 은전한닢을 이용해서 전처리하기로 결정.
<br>

```python
# 형태소 분석
# 꼬꼬마, 코모란, 트위터, 은전한닢
from konlpy.tag import Kkma, Komoran, Okt
from eunjeon import Mecab

kkma = Kkma()
kom = Komoran()
ok = Okt()
me = Mecab()

sentence = '아버지가방에들어가신다'

print(kkma.pos(sentence))
print(kom.pos(sentence))
print(ok.pos(sentence))
print(me.pos(sentence))
```

# 전처리과정<br><br>

#### 한국어 위키피디아를 형태소 분석하여 txt 형태로 저장
<br>

```python
# 은전한닢으로 형태소 분석하여 전처리
# 5개만 test
from eunjeon import Mecab

out_f = './data/prepro_ko_wiki_sample.txt'
in_f = './data/ko_wiki.txt'

me = Mecab()
# cp949 에러 해결을 위해 encoding='utf-8' 추가
output = open(out_f, 'wt', encoding='utf-8')

with open(in_f, 'r', encoding='utf-8') as rf:
    lines = rf.readlines()
    i = 0
    
    for line in lines:
        temp_arr = me.morphs(line)
        line = bytes(' '.join(temp_arr), 'utf-8').decode('utf-8') + '\n'
        output.write(line)
        
        print(line)
        i = i + 1
        if i == 5: break
        if i % 10000 == 0:
            print('Preprocessed ' + str(i) + ' articles')
    output.close()
    print('Preprocessing complete!')
```
